```python
html = '/n'.join(open('jianfan.html').readlines())
from bs4 import BeautifulSoup
soup = BeautifulSoup(html, 'lxml')
```

```python
ul = soup.select('ul')[3]
lis = ul.select('li')
fan = ''
jian = ''
for li in lis:
    jianfan = li.text.split(' → ')
    if jianfan[0] != jianfan[1]:
        fan += jianfan[0]
        jian += jianfan[1]

print(jian)
print(fan)
```
